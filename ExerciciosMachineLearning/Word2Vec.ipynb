```python
#Nltk tem metodos de manipução de texto
import nltk
#gensim tem varios modelos de analise de texto e embedings(Sendo um deles word2vec)
from gensim.models import Word2Vec
from nltk.corpus import stopwords
```

```python
#texto de treino

text = """A Guerra é um conflito armado[a] entre estados, governos, sociedades ou grupos paramilitares, como mercenários, insurgentes e milícias. Geralmente é caracterizada por extrema violência, agressão, destruição e mortalidade, usando forças militares regulares ou irregulares. A guerra refere-se às atividades e características comuns dos tipos de guerra, ou das guerras em geral.[2] Guerra total é uma guerra que não se restringe a alvos militares puramente legítimos e pode resultar em sofrimento e baixas em massa de civis ou outros não-combatentes.

Enquanto alguns estudiosos da guerra consideram a guerra um aspecto universal e ancestral da natureza humana,[3] outros argumentam que é resultado de circunstâncias socioculturais, econômicas ou ecológicas específicas.[4]

A guerra pode acontecer entre países ou entre grupos menores como tribos ou facções políticas dentro do mesmo país (confronto interno). Em ambos os casos, pode-se ter a oposição dos grupos rivais isoladamente ou em conjunto. Neste último caso, tem-se a formação de aliança(s).

Podem ser consideradas como algumas das maiores guerras travadas pela humanidade a Primeira Guerra Mundial e a Segunda Guerra Mundial."""
```

```python
#nltk.download('punkt')
```

```python
#tokenizando o texto
#transformando um texto que era uma grande string em uma lista de frases
sentences = nltk.sent_tokenize(text)

sentences
```

```python
#transformando cada frase em uma lista de palavras
#então irá acontecer uma tokenização de palavras das sentenças
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

sentences
```

```python
#nltk.download('stopwords')
```

```python
#demonstração das stopwords na nossa lingua
from nltk.corpus import stopwords

stopwords_portuguese = stopwords.words('portuguese')

stopwords_portuguese
```

```python
# remoção das stopwords
#a remoção ocorro por essas palavras não terem valor pra avaliação final


for i in range(len(sentences)):
    empty_sentence = []
    for word in sentences[i]:
        if word.lower() not in stopwords.words('Portuguese'):
            empty_sentence.append(word.lower())
        sentences[i] = empty_sentence
```

```python
#aqui são as palavras que ficaram pós remoção de stopwords
sentences
```

```python
#instanciando o Word2Vec e passando as sentenças para ele
model = Word2Vec(sentences, min_count=1)

words_vector = model.wv
```

```python
#checa quais vetores de embeding estão mais proximos do embeding da palavra escolhida
words_vector.most_similar("guerra")
```

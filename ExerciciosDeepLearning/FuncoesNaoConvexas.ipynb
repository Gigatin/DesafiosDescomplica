```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Desabilitando a execução imediata
tf.compat.v1.disable_eager_execution()

# Um vetor constante de tamanho 6
a = tf.constant([1.0, -0.5, 3.4, -2.1, 0.0, -6.5], dtype=tf.float32)

# Aplicando a função ReLU e armazenando o resultado em 'b'
b = tf.nn.relu(a, name='ReLU')

# Iniciando uma sessão do Tensorflow
with tf.compat.v1.Session() as sess:
    # Executar a sessão para calcular os valores de 'a' e 'b'
    a_val, b_val = sess.run([a, b])
    
    print('Tipo de entrada:', a)
    print('Entrada:', a_val)
    print('Tipo de retorno:', b)
    print('Saída:', b_val)
    
    # Plotar o resultado
    plt.plot(b_val)
    plt.grid()
    plt.show()


```

```python
#analisando dados
df = pd.read_csv('C:/csv/paises_do_mundo.csv')
df.head()
```

```python
#filtro pra ver paises com a letra k (checando corea do sul)
#strip pra tirar espaços vazios
df[df.pais.str.strip().str.contains('K')]
```

```python
#selecionando paises
paises = ['Angola', 'Argentina', 'Australia', 'Brazil', 'Chile',
          'Japan', 'Morocco', 'Senegal', 'Korea, South']

#base que vai receber paises
mini_df = df[df.pais.str.strip().isin(paises)].copy()

#filtrando colunas 4 a 10
X = mini_df.iloc[:, 4:10].dropna().copy()
```

```python
#Y = base a ser agrupada
#method = calculo de distancia entre os clusters
#metric = qual será a metrica de distancia? (Checar as distancias em scipy pdist)
Z = linkage(y=X, method='ward', metric='euclidean')

plt.figure(figsize=(7,5))#tamanho 7 por 5
dendrogram(Z, labels=list(mini_df.pais),#visualizando o dendograma, z=agrupamento, para ver os nomes do país pega o mini_df coluna país
           orientation='top',#parametro de orientação para ver de cima pra baixo
           leaf_font_size=10,leaf_rotation=60,);#definindo leitura de tamanho 10 com a fonte em diagonal pela rotação em 60
```

```python
#testando com metodo single pra checar a diferença com o ward
Z = linkage(X, method='single',metric=)

plt.figure(figsize=(7,5))
dendrogram(Z, labels=list(mini_df.pais),
           orientation='top',
           leaf_font_size=10,leaf_rotation=60,);
#Korea do sul ta mais distante da america do sul
```

```python
#metodo centroid, valor medio de cada grupo pra considerar a distancia de um grupo pro outro
Z = linkage(X, method='centroid')

plt.figure(figsize=(7,5))
dendrogram(Z, labels=list(mini_df.pais),
           orientation='top',
           leaf_font_size=10,leaf_rotation=60,);

#nesse caso os clusters ficam divididos entre paises africanos, da america do sul e asiativos
```

```python
#Hierarquica na pratica:

import pandas as pd
import numpy as np

import seaborn as sns
sns.set_theme(style="darkgrid", palette="pastel")
from matplotlib import pyplot as plt

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
#checando base de dados
data = pd.read_csv('C:/csv/tamanho_pinguins.csv')
data.head()
```

```python
#apagando valores faltantes
data = data.dropna()

#remoção de outiliers (Bicos maiores que 50)
data = data[data.larg_bico_mm < 50]
```

```python
#333 linhas e 7 colunas
data.shape
```

```python
# x = valores numericos, y = especie
X, y = data.iloc[:, 2:-1], data.iloc[:, 0]

#rescalando dados
scaler = MinMaxScaler()
X_ = scaler.fit_transform(X)
X_
```

```python
#rodar o k-means só com 2 variaveis (Comprimento asa e bico)
x = X_[:, [0,2]]
kmeans = KMeans(n_clusters=3, random_state=3, n_init="auto").fit(x)
labels = kmeans.labels_

plt.scatter(x[labels == 0, 0], x[labels == 0, 1], color="green", edgecolors='black', s=15)
plt.scatter(x[labels == 1, 0], x[labels == 1, 1], color="red", edgecolors='black', s=15)
plt.scatter(x[labels == 2, 0], x[labels == 2, 1], color="blue", edgecolors='black', s=15)
plt.xlabel('comp_bico_mm')
plt.ylabel('comp_asa_mm');
```

```python
#filtrando só bico e asa, definindo clusters
df_res = pd.DataFrame(x, columns = ['comp_bico_mm', 'comp_asa_mm'])
df_res['cluster'] = labels
```

```python
df_res
```

```python
#dentro dos clusters foi encontrado 3 grupos diferentes
sns.scatterplot(data=df_res, 
                x="comp_bico_mm", 
                y="comp_asa_mm", 
                hue="cluster", 
                palette="deep");
```

```python
#distancia entre os pinguins
Z = linkage(X_, method='centroid')

plt.figure(figsize=(12,5))
dendrogram(Z,  truncate_mode='lastp', labels=list(data.especie),#se remover o lastp a imagem fica bem grande mas muito bagunçada
           orientation='top',
           leaf_font_size=10,leaf_rotation=60,);
```
